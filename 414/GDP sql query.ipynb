```python
#connect to db
con = psycopg2.connect(
        host = "localhost",
        database="postgres",
        user="postgres",
        password="allow",
        port=5432
)

#cursor
cur=con.cursor()
#CREATE TABLE
create_table = """                  
        create table mytest.gdp(
              code varchar NOT NULL, 
              name varchar NOT NULL, 
              year bigint NOT NULL, 
              year1 bigint NOT NULL, 
              year2 bigint NOT NULL, 
              year3 bigint NOT NULL, 
              year4 bigint NOT NULL, 
              year5 bigint NOT NULL, 
              year6 bigint NOT NULL, 
              year7 bigint NOT NULL, 
              year8 bigint NOT NULL, 
              year9 bigint NOT NULL
         )
     """ #execute query
cur.execute(create_table)

#add values to table (PUT)
def input_values(data):
        try:
                for (index, row) in data.iterrows():
                        insert_sql_query = f'''
                                insert into mytest.gdp (Code, Name, year, year1, year2, year3, year4, year5, year6, year7, year8, year9)
                                values('{row['Code']}'
                                , '{row['Name'].replace("'", " ")}'
                                , '{row['2010']}'
                                , '{row['2011']}'
                                , '{row['2012']}'
                                , '{row['2013']}'
                                , '{row['2014']}'
                                , '{row['2015']}'
                                , '{row['2016']}'
                                , '{row['2017']}'
                                , '{row['2018']}'
                                , '{row['2019']}'
                                )'''

                        cur.execute(insert_sql_query)
        except psycopg2.errors.UniqueViolation:
                print('data already exist, add new data')
                pass
data_csv = pandas.read_csv('GDP.csv')
input_values(data_csv)

#MODIFY
cur.execute("UPDATE mytest.gdp SET year=10504 WHERE Code='ZWE'")
cur.execute("UPDATE mytest.gdp SET year2=5043 WHERE Code='AGO'")
cur.execute("UPDATE mytest.gdp SET year1=10005 WHERE Code='EGY'")
print('Modified items are moved to end of table')

#FETCH
cur.execute("SELECT * from mytest.gdp")
rows=cur.fetchall()
for r in rows:
        print(r)
#DELETE TABLE
cur.execute("DROP TABLE mytest.gdp")
#to apply change
con.commit()
#close crusor
cur.close()
#close connection
con.close()
```
